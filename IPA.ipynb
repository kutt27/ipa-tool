```python
import nltk
nltk.download('cmudict')
```

```python
from nltk.corpus import cmudict

d = cmudict.dict()

ARPABET_TO_IPA = {
    'AA': 'ɑ', 'AE': 'æ', 'AH': 'ʌ', 'AO': 'ɔ', 'AW': 'aʊ', 'AX': 'ə',
    'AY': 'aɪ', 'EH': 'ɛ', 'ER': 'ɜr', 'EY': 'eɪ', 'IH': 'ɪ', 'IX': 'ɨ',
    'IY': 'i', 'OW': 'oʊ', 'OY': 'ɔɪ', 'UH': 'ʊ', 'UW': 'u', 'UX': 'ʉ',
    'B': 'b', 'CH': 'tʃ', 'D': 'd', 'DH': 'ð', 'EL': 'l̩', 'EM': 'm̩',
    'EN': 'n̩', 'F': 'f', 'G': 'ɡ', 'HH': 'h', 'JH': 'dʒ', 'K': 'k',
    'L': 'l', 'M': 'm', 'N': 'n', 'NX': 'ɾ̃', 'NG': 'ŋ', 'P': 'p',
    'Q': 'ʔ', 'R': 'ɹ', 'S': 's', 'SH': 'ʃ', 'T': 't', 'TH': 'θ',
    'V': 'v', 'W': 'w', 'WH': 'ʍ', 'Y': 'j', 'Z': 'z', 'ZH': 'ʒ',
    'DX': 'ɾ'
}
```

```python
def analyze_word(word):
    """
    Looks up a word in CMUdict, counts syllables, and converts to IPA.
    """
    word = word.lower().strip(".,!?;:()\"'")

    if word not in d:
        return {
            "word": word,
            "ipa": "[Unknown]",
            "syllables": 0,
            "error": "Word not found in dictionary"
        }

    # Get the first phonetic pronunciation variant from the dictionary
    phonemes = d[word][0]

    syllable_count = 0
    ipa_phones = []

    for phone in phonemes:
        # If the phoneme ends with a digit (0, 1, 2), it's a vowel/syllable kernel
        if phone[-1].isdigit():
            syllable_count += 1
            # Strip the stress number to get the clean Arpabet token
            clean_phone = phone[:-1]
        else:
            clean_phone = phone

        # Translate to IPA, fallback to original if missing
        ipa_symbol = ARPABET_TO_IPA.get(clean_phone, clean_phone)
        ipa_phones.append(ipa_symbol)

    # Join symbols together; slashes denote phonetic transcription
    ipa_transcription = f"/{''.join(ipa_phones)}/"

    return {
        "word": word,
        "ipa": ipa_transcription,
        "syllables": syllable_count
    }
```

```python
sentence = "Natural language processing is incredibly fun"

print(f"{'WORD':<15} | {'IPA':<15} | {'SYLLABLES'}")
print("-" * 45)

for word in sentence.split():
    result = analyze_word(word)
    print(f"{result['word']:<15} | {result['ipa']:<15} | {result['syllables']}")
```
